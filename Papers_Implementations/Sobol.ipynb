## Abstract 
Global sensitivity indices for rather complex mathematical models can be efficiently computed by Monte Carlo
(or quasi-Monte Carlo) methods. These indices are used for estimating the influence of individual variables or groups of variables on the model output.

## Introduction

$u = f(x)$

$x = (x_1,...,x_n)$

$n$: number of dimentions

Local Sensibility: $\Big(\frac{\partial u}{\partial x}\Big)_{x=x^*}$

## AVOVA-representation
$D=\int f^2dx - f_0^2$ , $D_{i_1 ... i_s}=\int f^2dx_{i_1} ... dx_{i_s} $

$D=\sum_{s1=1}^{n} \sum_{i_1<...<i_s}^{n} D_{i_1 ... i_s}$

$n$: number of dimentions

## Sensibility indeces

Definition 2: the ratios
$S_{i_1 ... i_s} = \frac{D{i_1 ... i_s}}{D}$ are called global **sensitivity indeces**.

$\sum_{s1=1}^{n} \sum_{i_1<...<i_s}^{n} S_{i_1 ... i_s}=1$
