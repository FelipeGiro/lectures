## Task 1

Consider a component of offshore wind turbine subjected to a tensile fatigue loading, which can be represented in terms of a normal distributed equivalent stress range S with parameters ðœ‡s = 30 MPa, covs = 0.15. The component is estimated to experience 105 cycles per year of the given stress range.
Imperfections are present during the manufacturing process of the steel bar and associating initial defect A0 can be modelled as an exponential distributed random variable with ðœ‡A0 = ÏƒA0 = 0.5 mm. The crack growth due to the stress cycles can be simply modelled by:

A(n) = A_0*exp(C*pi*S**2*n)

Where C is a material parameter equal to 5*10-10
. The thickness of the steel bar is 20 mm, and it is assumed that the steel bar fails when the crack size has reached its thickness. Based on the simple crack growth model, calculate the annual probability of failure over the lifetime of 20 years using MCS method.

```python
samples = int(1e6)

# stress range [MPa]
S = np.random.normal(30, 0.15*30, size=samples)

# number of cycles per year
n = 1e5

# intial crack size
A_0 = np.random.exponential(0.5, size=samples)

# material parameter
C = 5e-10

# bar thickness [mm]
t = 20

# lifetime [years]
T = 20
```

```python
# cycles for 20 years
n_20 = n*T

# crack size after 20 years
A = A_0*np.exp(C*np.pi*S**2*n_20)

n_failure = np.sum(A > t)
n_total   = len(A)

# probability of failure
p_f = n_failure/n_total

print('The probability of failure is', p_f)
```

```python
# annual probability
years = np.linspace(0,T,num=T+1)
n_array = years*1e5
cum_pf_annual = []

for n_year in n_array:
    A = A_0*np.exp(C*np.pi*S**2*n_year)
    cum_pf_annual += [np.sum(A > t)/len(A)]

pf_annual = np.diff(cum_pf_annual)

plt.title('Annual probability of failure')
plt.plot(years[1:], pf_annual)
plt.xlabel('years')
plt.ylabel('p_f')
plt.grid()
plt.show()
```
